```python
import pandas as pd
import barnum
```

```python
numTimes = 100
zip_codes = []

for counter in range(numTimes):
    zip_codes.append(barnum.create_city_state_zip()[0])
    
print(zip_codes)
```

```python
zip_code_list = pd.DataFrame(zip_codes)

zip_code_list.columns = ["Zip Codes"]
zip_code_list
```

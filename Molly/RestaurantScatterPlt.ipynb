```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 24})
import pandas as pd
import numpy as np

from matplotlib import pyplot as mp
```

```python
merge_info = "MergedData.csv"
merged_df = pd.read_csv(merge_info)

merged_df.head()
```

```python
merged_df = merged_df.rename(columns={'Percent of households with income over $200,000': '% of households with income over $200,000', 'Total GF Rest': 'Total Gluten-Free Restaurants'})
```

```python
# Create a scatter plot which compares Percent of household ___  Total GF Rest
merged_df.plot(kind="scatter", x="% of households with income over $200,000", y="Total Gluten-Free Restaurants", grid=True, figsize=(30,10), s=80, color='orange',
              title="Wealth v. Total # of GF Restaurants")

mp.savefig('Wealth_v_TotalGF.png')
plt.show()
```

```python
final_info = "final_merge.csv"
final_df = pd.read_csv(final_info)

final_df.head()
```

```python
final_df = final_df.rename(columns={'Percent of households with income over $200,000': '% of households with income over $200,000', 'ALL Total': 'Total Restaurants'})
```

```python
final_df.plot(kind="scatter", x="% of households with income over $200,000", y="Total Restaurants", grid=True, figsize=(30,10), marker='D', s=60, color='magenta',
              title = "Wealth v. Total # of Restaurants")

mp.savefig('Wealth_v_TotalRest.png')
plt.show()
```
